#  Image Processing

### Blending the images and pasting the images

```python
import cv2
```

```python
img1 = cv2.imread('dog_backpack.png')
img1 = cv2.cvtColor(img1 , cv2.COLOR_BGR2RGB)
img2 = cv2.imread('watermark_no_copy.png')
img2 = cv2.cvtColor(img2 , cv2.COLOR_BGR2RGB)
```

```python
import matplotlib.pyplot as plt
```

```python
# displaying the image
plt.imshow(img1)
plt.show()
```

```python
# getting the second image
plt.imshow(img2)
plt.show()
```

```python
# the shape of the image 
img1.shape
```

```python
img2.shape
```

```python
# Blending the images
img1 = cv2.resize(img1 , (1200 ,1200))
img2 = cv2.resize(img2 , (1200 , 1500))
```

```python
plt.imshow(img1)
plt.show()
```

```python
plt.imshow(img2)
plt.show()
```

```python
# from this what we leaned is we can resize the image or blend the image using the cv2
```

```python
# addWeighted(src1, alpha, src2, beta, gamma[, dst[, dtype]])
blend = cv2.addWeighted(src1 = img1 , alpha = 0.5 , src2 = img2 , beta = 0.5 , gamma = 0)
```

```python
plt.imshow(blend)
plt.show()
```

```python
blend = cv2.addWeighted(src1 = img1 , alpha = 0.8 , src2 = img2 , beta = 0.1 , gamma = 0)
```

```python
plt.imshow(blend)
```

# Overlay small image on top of a large image (No Blending )

### Numpy Reassignment

```python
img1 = cv2.imread('dog_backpack.png')
img1 = cv2.cvtColor(img1 , cv2.COLOR_BGR2RGB)
img2 = cv2.imread('watermark_no_copy.png')
img2 = cv2.cvtColor(img2 , cv2.COLOR_BGR2RGB)
```

```python
img2 = cv2.resize(img2, (600,600))
```

```python
plt.imshow(img2)
```

```python
plt.imshow(img1)
```

```python
# take small image and large image

large_img = img1
small_img = img2
```

```python
x_offset = 0
y_offset = 0
```

```python
x_end = x_offset + small_img.shape[1]
y_end = x_offset + small_img.shape[0]
```

```python
large_img[y_offset: y_end, x_offset: x_end] = small_img
```

```python
plt.imshow(large_img)
```
